```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset (assumes train.csv is already uploaded)
df = pd.read_csv("train.csv")

# Data Cleaning
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop(columns=['Cabin'], inplace=True)

# Set plot style
sns.set(style="whitegrid")

# Create figure and axes
fig, axs = plt.subplots(3, 2, figsize=(15, 14))  # 3 rows x 2 columns layout

# Plot 1: Survival Count
sns.countplot(data=df, x='Survived', palette='Set2', ax=axs[0, 0])
axs[0, 0].set_title("Survival Count", fontweight='bold')

# Plot 2: Survival by Gender
sns.countplot(data=df, x='Sex', hue='Survived', palette='pastel', ax=axs[0, 1])
axs[0, 1].set_title("Survival by Gender", fontweight='bold')

# Plot 3: Age Distribution
sns.histplot(df['Age'], bins=30, kde=True, color='skyblue', ax=axs[1, 0])
axs[1, 0].set_title("Age Distribution", fontweight='bold')

# Plot 4: Survival by Class
sns.countplot(data=df, x='Pclass', hue='Survived', palette='Set3', ax=axs[1, 1])
axs[1, 1].set_title("Survival by Class", fontweight='bold')

# Plot 5: Embarked vs Survival
sns.countplot(data=df, x='Embarked', hue='Survived', palette='coolwarm', ax=axs[2, 0])
axs[2, 0].set_title("Survival by Embarkation Port", fontweight='bold')

# Hide unused subplot (bottom right corner)
axs[2, 1].axis('off')

# Adjust layout
plt.tight_layout()
plt.show()
```
